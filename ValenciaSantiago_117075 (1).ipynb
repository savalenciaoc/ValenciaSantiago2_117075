```python
#Ejercicio_16 - Lista de números ordenados

import sys #Importa el modulo sys

print('Programa para agregar un número a una lista ascendente')

#Guarda una lista con los números
Lista = eval(input('Ingrese una lista de números ascendente: '))
x = len(Lista) #Guarda el tamaño de la lista
cont = 0 #Contador

for i in range(0,x-1): #Itera el tamaño de la lista-1
    if Lista[i] <= Lista[i+1]: #Condición de si el elemento i de la lista
                               #es menor o igual que el siguiente
        cont = cont 
    else: #Condición de si el elemento i de la lista es mayor que el siguiente
        cont += 1 #Aumenta 1 el contador
if cont !=0: #Condición de si el contador es distinto de cero
    sys.exit('La lista no está en orden ascendente') #Si el contador es 
                                                     #distinto de cero la lista
                                                     #no está en orden
                                                     #ascendente

num = int(input('Ingrese un número: ')) #Se guarda el número que agregaremos a 
                                        #la lista
    
for i in range(0,x): #Itera en un rango del tamaño de la cadena
    if num>=Lista[-1]:  #Si el último dígito de la lista es menor que
                        #num se adicionará el número a la lista y se
                        #romperá el código. De lo contrario, se irá
                        #iterando por índice hasta que se encuentre un
                        #número que sea menor o igual
        Lista.append(num) #Se agrega el número a la lista en la última posición
        break
    if num<=Lista[i]: #Si el número es menor que el elemento i de la lista
        Lista.insert(i,num) #Se incerta el número en la posición i
        break

print(Lista) #Imprime la lista con el número añadido
```
